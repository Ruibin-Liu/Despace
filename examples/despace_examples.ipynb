## Imports

```python
from despace import SortND
import numpy as np
```

## Usage
`SortND` is a python class for sorting N-D array-like data and visualize the sorted data if the number of dimensions is <=3.

### 1. 2D case 

```python
data = np.random.rand(100, 2)  # Let's generate some random data
```

We can either initiate the `SortND` class with data and then use the `sort` method, or initiate the `SorND` class withtou providing data, but then directly call with data.

```python
s = SortND(data)
s.sort()
s.plot(show_plot=True, plot_arrow=True)
```

The data is colored from blue to red as the index goes higher. The arrows coonect two successive data points.

If we don't want the arrows that show the relative indices in the sorted data, we can set `plot_arrow` to False

```python
s.plot(show_plot=True, plot_arrow=False)
```

Let's look at the sorted data:

```python
s.sorted_coords
```

### 2. 3D case

```python
data = np.random.rand(1000, 3)
s(data)
s.plot(show_plot=True)
```

### 3. Try some organized data

```python
data = [[i, j] for i in range(8) for j in range(8)]
s.sort(data)
s.plot(show_plot=True, plot_arrow=True)
```

## We can see we get the Morton space filling curve!

If one dimension length is odd:

```python
data = [[i, j] for i in range(8) for j in range(9)]
s.sort(data)
s.plot(show_plot=True, plot_arrow=True)
```

The path can cross itself in the 'odd' dimension.

If both dimension lengths are odd:

```python
data = [[i, j] for i in range(9) for j in range(9)]
s.sort(data)
s.plot(show_plot=True, plot_arrow=True)
```

Since the order is random if two points have the same value in either of their components, we see move data cross-over in both directions!

If we set `sort_type='Hilbert'`:

```python
s.sort(data, sort_type='Hilbert')
s.plot(show_plot=True, plot_arrow=True)
```

## We get the Hilbert space filling curve (2D)!
